```python
def computecost(X_train,y_train,weights):  # z=np.power(((X_train*weights.T)-y_train),2)
    #weights = np.matrix(np.array(weights))
    z=np.power(((np.dot(X_train,weights.T))-y_train),2)
    return np.sum(z)/(len(X_train))

print('cost is:',computecost(X_train,y_train,weights))
```

```python
def leastsquaremethod(X_train,y_train):
    X_train=np.array(X_train)
    y_train=np.array(y_train)
    W=np.linalg.inv(X_train.T.dot(X_train)).dot(X_train.T.dot(y_train))
    return W

print("least square method weights: ",leastsquaremethod(X_train,y_train))
    
```

```python
 

def gradient_descent(X_train,y_train,weights,alpha,iters):
    #weights = np.matrix(np.array(weights))
    m=X_train.shape[0]
    costhistory=[]
    accuracyhistory = [] 
    weightshistory=[]
    for i in range(iters):
        y_predict=X_train.dot(weights.T)
        #cost=y_predict-y_train
        costvalue=computecost(X_train,y_train,weights)
        difference=X_train.T.dot(X_train.dot(weights.T)-y_train)
        weights=weights-(alpha*2/m)*difference.T
        weights=np.array(weights).reshape((1,4))
        costvalue=computecost(X_train,y_train,weights)
        accuracyvalue = 1-costvalue
        accuracyhistory.append(accuracyhistory)
        costhistory.append(costvalue)
        weightshistory.append(weights)
        if (i%1000 == 0):
            print(f"iter: {i} , w:  {weights} , cost:  {costvalue} , Acc:  {accuracyvalue*100}\n")
            print("cost after modify:",costvalue)
    return weightshistory,costhistory,accuracyhistory,weights

alpha=0.001
iters=100000
    

weightshistory,cost,accuracy,weights=gradient_descent(X_train,y_train,weights,alpha,iters)
weights
# print("cost is",cost)
print('final weights =',weights)
#print('Cost after modify =',computecost(X_train,y_train,weights))
```
